### **1. Distribuzioni di probabilità**
Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione.

Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

Stima il parametro di forma dal campione usando il metodo fit.

Extra:
Le distribuzioni hanno molti metodi utili. Esplorali usando il completamento automatico con il tasto TAB.

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
gamm=sp.stats.gamma(1)
distribuzione=gamm.rvs(size=1000)
plt.hist(distribuzione,bins=50,density=True)
plt.plot(np.sort(distribuzione),gamm.pdf(np.sort(distribuzione)))
plt.plot(np.sort(distribuzione),gamm.cdf(np.sort(distribuzione)))
```

```python
fit=sp.stats.gamma.fit(distribuzione)
print(fit[0])
gamm.var()
```

### **2. Fitta i dati**
Prova a fittare i dati sottostanti con le migliori curve, calcola il MAE e l'RMSE
Esercizio 

```python
import numpy as np
temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
import matplotlib.pyplot as plt
def lin(x,m,q):
    return m*x+q

```

```python
plt.figure(figsize=(8,5))
plt.scatter(temp_max,temp_min)
plot,_=sp.optimize.curve_fit(lin,temp_max,temp_min)
plt.plot(temp_max,lin(temp_max,plot[0],plot[1]))
```

```python
print(np.mean(abs(lin(temp_max,plot[0],plot[1])-temp_min)))
print(np.sqrt(np.mean((lin(temp_max,plot[0],plot[1])-temp_min)*(lin(temp_max,plot[0],plot[1])-temp_min))))
```
